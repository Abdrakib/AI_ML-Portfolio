```python
import pandas as pd
import numpy as np
import neattext.functions as nfx
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('udemy_course_data.csv')
df.head()
```

## We will work on course title because these are the course name we will use to give recomendation

```python
# clean stop word from the title and remove special character
df['Clean_title'] = df['course_title'].apply(nfx.remove_stopwords)
df['Clean_title'] = df['Clean_title'].apply(nfx.remove_special_characters)
df.head()
```

## vectorize the clean title

```python
countvect = CountVectorizer()
cvmat = countvect.fit_transform(df['Clean_title'])
cvmat 
```

## cosine similarity

```python
cos_sin = cosine_similarity(cvmat)
cos_sin
```

## Recommend Course

```python
title = 'How To Maximize Your Profits Trading Options'
```

```python
course_index = pd.Series(
    df.index, index = df['course_title']).drop_duplicates()
course_index[title]
```

```python
index = course_index[title]
```

```python
list(enumerate(cos_sin[index]))
```

```python
course_index = pd.Series(
    df.index, index = df['course_title']).drop_duplicates()
index = course_index[title]
scores = list(enumerate(cos_sin[index]))
sorted_scores = sorted(scores, key = lambda x: x[1], reverse = True)
selected_course_index = [i[0] for i in sorted_scores[1:]]
selected_course_score = [i[1] for i in sorted_scores[1:]]
rec_df = df.iloc[selected_course_index]
rec_df['Similarity_Score'] = selected_course_score
final_recommended_courses = rec_df[[
    'course_title', 'Similarity_Score', 'url', 'price', 'num_subscribers']]
final_recommended_courses
```
